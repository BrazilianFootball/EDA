```python
df = pd.read_csv('https://www.football-data.co.uk/new/BRA.csv').dropna()
df = df[df['Season'] >= 2013].reset_index(drop = True)
df['Home'] = df['Home'].apply(rename)
df['Away'] = df['Away'].apply(rename)
df['HG'] = df['HG'].astype(int)
df['AG'] = df['AG'].astype(int)
df = df[[col for col in df.columns if 'Max' not in col and 'Avg' not in col]]
cols = df.columns
df['IPH'] = 1 / df['PH']
df['IPD'] = 1 / df['PD']
df['IPA'] = 1 / df['PA']
df['Total'] = df['IPH'] + df['IPD'] + df['IPA']
df['IPH'] = df['IPH'] / df['Total']
df['IPD'] = df['IPD'] / df['Total']
df['IPA'] = df['IPA'] / df['Total']
df['H'] = (df['Res'] == 'H') * 1
df['D'] = (df['Res'] == 'D') * 1
df['A'] = (df['Res'] == 'A') * 1
df['De Finetti'] = (df['IPH'] - df['H']) ** 2 + (df['IPD'] - df['D']) ** 2 + (df['IPA'] - df['A']) ** 2
print(f'De Finetti: {df["De Finetti"].mean():.4f}')
df[cols]
```
